```python
# Seaborn for plotting and styling
import seaborn as sns

# importing datasets (dataframe)
df = sns.load_dataset('tips')
print (df.head()) #top 5 rows
```

```python
df.smoker.unique()
#df["smoker"].unique()
```

```python
df.day.value_counts()
```

```python
df.time.value_counts()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df)

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="day")

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="smoker")

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="time")

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="time", hue_order=["Dinner", "Lunch"])

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="sex")

plt.show()
```

```python
hue_colors = {"Yes": "black",
"No": "red"}

sns.scatterplot(x="total_bill",
y="tip",
data=df,
hue="smoker",hue_order=["Yes","No"], palette=hue_colors)

plt.show()
```

```python
#countplot on categorical variables (univariate)
sns.countplot(x="smoker",
data=df, hue="sex")

plt.show()
```

```python
#countplot on categorical variables (univariate)
sns.countplot(x="time",
data=df, hue="day")

plt.show()
```

```python
#countplot on categorical variables (univariate)
sns.countplot(x="sex",
data=df,hue="day")

plt.show()
```
