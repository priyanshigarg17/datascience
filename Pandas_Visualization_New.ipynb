```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.plotting.backend = 'plotly'
```

```python
data = {'Sales': [1000, 1500, 1200, 1800, 2000],
    'Expenses': [600, 700, 800, 1000, 1200]}

Year = ["2018", "2019", "2020", "2021", "2022"]

df = pd.DataFrame(data,index=Year)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)
#histogram will convert your data into class intervals: x axis
#y axis will show the frequency
#univariate plot

plt.hist(x, bins=10, histtype="stepfilled")#bins represent class interval
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#You can create a stratified boxplot using the by keyword argument to create groupings.
#For instance,
df = pd.DataFrame(np.random.rand(10, 2), columns=["Col1", "Col2"])

df["X"] = pd.Series(["M", "M", "M", "M", "M", "F", "F", "F", "F", "F"])
df
```

```python
df.boxplot(by="X")
```

```python
#You can also pass a subset of columns to plot, as well as group by multiple columns:

df = pd.DataFrame(np.random.rand(10, 3), columns=["Col1", "Col2", "Col3"])

df["X"] = pd.Series(["M", "M", "M", "M", "M", "F", "F", "F", "F", "F"])

df["Y"] = pd.Series(["Y", "Y", "N", "Y", "N", "N", "Y", "Y", "N", "Y"])

df
```

```python
df.boxplot(column=["Col1","Col2"],by=["X","Y"])
```

```python
#variations in scatter plot

df = pd.DataFrame(np.random.rand(50, 4), columns=["a", "b", "c", "d"])

df["species"] = pd.Categorical(
    [0] * 20 + [1] * 20 + [2] * 10
)
df
```

```python
df.plot(x="a",y="b",kind="scatter",c="species",s=20,colormap="rainbow")
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal="kde")
```
